```python
import pandas as pd 
import numpy as np
import glob
import os
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches
```

```python
df1 = pd.read_csv('/media/DataD800/Alina/main_set/MLtables_cope/demographics_table_new.csv', index_col=0)
```

```python
df2 = pd.read_csv('/media/data/HCPAging/data/MLTablesMultCon/demography.csv', index_col=0)
```

```python
df3 = pd.read_csv('/media/data/Dunedin_Study_Data_Narun_P_Jan2022/MLTabs/info.csv', index_col=0)
```

```python
age1 = df1['Age_in_Yrs']
```

```python
age2 = (df2['interview_age']/12).round(0).astype(int)
```

```python
age3 = pd.Series(np.full(len(df3.index), 45), index=df3.index )
```

```python
vec1 = pd.concat([age1, age2, age3], axis=0)

vec2 = pd.concat([pd.Series(np.full(len(age1.index), 'HCP-YA'), index = age1.index), 
                  pd.Series(np.full(len(age2.index), 'HCP-A'), index = age2.index), 
                  pd.Series(np.full(len(age3.index), 'DUD'), index = age3.index)], axis=0)

dset = pd.DataFrame({'Age':vec1,'group':vec2})
```

```python
plt.figure(figsize=(20,15))

sns.histplot(data = dset, x='Age', hue='group', bins=25, legend=False, palette='bright')

plt.xticks(fontsize=60)  
plt.yticks(fontsize=60)
plt.xlabel('Age',fontsize=75)
plt.ylabel('',fontsize=60)

plt.title('Number of participants', fontsize=85)

plt.legend(handles=[mpatches.Patch(color=sns.color_palette("bright")[0], label='HCP Young \nAdults (n=873)\n\n'),
                    mpatches.Patch(color=sns.color_palette("bright")[1], label='HCP Aging \n(n=504)\n\n'),
                    mpatches.Patch(color=sns.color_palette("bright")[2], label='Dunedin Study \n(n=754)\n\n')],
           loc='upper right', fontsize=50) #bbox_to_anchor=(1.18, 1.015), 

plt.show()
```
